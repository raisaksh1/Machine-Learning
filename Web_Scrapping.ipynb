```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpages = requests.get('https://www.amazon.in/s?k=shoes+for+women&page=2&xpid=2sBXn9SybG002&crid=UWYYG0Z8G995&qid=1714571759&sprefix=shoes%2Caps%2C239&ref=sr_pg_2', headers = headers).text
```

```python
soup = BeautifulSoup(webpages, 'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h2'):
  print(i.text)
```

```python
len(soup.find_all('a', class_ = 'a-popover-trigger a-declarative'))
```

```python
company = soup.find_all('div', class_ =  'a-section a-spacing-base a-text-center')
```

```python
len(company)
```

```python
name = []
price = []
d_date = []
rating = []
a_price = []

for i in company:
  name.append(i.find('h2').text)
  price.append(i.find('span', class_="a-price-whole").text)
  d_date.append(i.find('span', class_="a-color-base a-text-bold"))
  rating.append(i.find('i', class_= "a-icon a-icon-star-small a-star-small-4 aok-align-bottom"))
  a_price.append(i.find('span', {"aria-hidden" : "true"}).text)


# After scrapping of the little bit of the data what I have done is to create the dataframe out of it
d = {'name':name, 'price':price, 'actual_price':a_price, 'rating':rating}

df = pd.DataFrame(d)

df
```
