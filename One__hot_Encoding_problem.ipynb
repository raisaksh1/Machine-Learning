```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/carprices.csv")
df
```

```python
df.rename(columns = {'Car Model' : 'Models'}, inplace = True)
```

```python
dummies = pd.get_dummies(df.Models)
dummies
```

```python
merged = pd.concat([df, dummies], axis = 1)
merged
```

```python
mer = merged.drop(['Models', 'Mercedez Benz C class' ], axis = 1)
mer
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = mer.drop(['Sell Price($)'], axis = 1)
X
```

```python
m = mer.rename(columns = {'Sell Price($)': 'Prices'})
m
```

```python
Y = m.Prices
Y
```

```python
model.fit(X,Y)
```

```python
model.predict([[6000, 4, 0, 1]])
```

```python
model.score(X,Y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.Models = le.fit_transform(df.Models)
df
```

```python
X = df[['Models', 'Mileage']]
X
```

```python
Y = df[['Sell Price($)']]
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("Models", OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)
```

```python
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,Y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.score(X,Y)
```
