```python
import pandas as pd
path = '/content/model3.xlsx'
df = pd.read_excel(path)
df
```

```python
dummies = pd.get_dummies(df.Mileage)
dummies
```

```python
merged = pd.concat([df , dummies], axis = 'columns')
merged
```

```python
merged.columns
```

```python
final = merged.drop(['Mileage','udi'], axis ='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('Age', axis ='columns')
X
```

```python
y = final.Age
y
```

```python
model.fit(X,y)
```

```python
model.predict([[68000,0,1]])
```

```python
model.predict([[72000,0,0]])
```

```python
model.score(X,y)
```
